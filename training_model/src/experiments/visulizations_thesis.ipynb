```python
import os
import sys
import pandas as pd
import numpy
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt 
sys.path.append(os.getenv('PATH_TO_TM'))
from src.dataset.dataset import Dataset
```

```python
PATH =  os.getenv('PATH_TO_TM')+ "/src/data/"
dataset = Dataset(filepath=PATH, subject='subject3')
dataset.preprocess_data(measurement_timeframe="3000ms")
```

```python
X, y = dataset.feature_extraction()
self_data = pd.DataFrame()
deepfake_data = pd.DataFrame()
for data, label in zip(X.values, y):
    data = pd.DataFrame(data={'max': [data[1]],'mean': [data[2]], 'str': [data[3]]})
    if label=='self':
        self_data = pd.concat([self_data, data])
    if label=='deepfake':
        deepfake_data = pd.concat([deepfake_data, data])
self_data['label'] = 'self'
deepfake_data['label'] = 'deepfake'
data_ = pd.concat([self_data, deepfake_data])
```

```python
fig = px.scatter(data_, x="mean", y="str", color="label")
fig.show()
```

```python
fig = px.scatter_3d(data_, x= 'max', y="mean", z="str", color="label", size_max=3)
fig.show()
```
